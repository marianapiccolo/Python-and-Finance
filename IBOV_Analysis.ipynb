# IBOV Analysis

Ibovespa Index (IBOV) is the most important indicator of the average performance of the quotations of shares traded on B3 - Brasil, Bolsa, Balcão. It is made up of the stocks with the highest volume traded in recent months.

Libraries:<br>
Pandas<br>
Pandas_datareader<br>
Numpy<br>
Matplotlib<br>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#%pip install pandas_datareader
import pandas_datareader.data as web
```

```python
ibov_price = web.DataReader('^BVSP', data_source = 'yahoo', start = '2020-01-01', end = '2020-11-10')
display(ibov_price)
```

# Graphical Analysis

```python
ibov_price['Adj Close'].plot(figsize = (15,5))
```

# IBOV Return

```python
ibov_return = ibov_price['Adj Close'][-1]/ibov_price['Adj Close'][0] - 1
print('IBOV Return: {:.2%}'.format(ibov_return))
```

# Moving Average

```python
ibov_price['Adj Close'].plot(figsize = (15,5), label = 'IBOV')
ibov_price['Adj Close'].rolling(21).mean().plot(label = 'Moving Average 21') # 21 days
ibov_price['Adj Close'].rolling(34).mean().plot(label = 'Moving Average 34') # 34 days
plt.legend()
plt.show()
```
